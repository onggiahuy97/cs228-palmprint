```python
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path

data_root = Path("datasets/Original")

# Collect one image per subject (left hand) for the first 10 subjects
samples = []
for subject_dir in sorted(data_root.glob("H_ID*"))[:10]:
    images = sorted((subject_dir / "L").glob("*.JPG"))
    if images:
        samples.append((images[0], subject_dir.name, "L"))

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
fig.suptitle("Sample Palmprint Images (Left Hand)", fontsize=14)

for ax, (img_path, subject, hand) in zip(axes.flat, samples):
    img = Image.open(img_path)
    ax.imshow(img)
    ax.set_title(f"{subject} {hand}", fontsize=9)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
